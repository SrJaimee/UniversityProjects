```python
#Caracteristicas (X) and etiquetas (Y)
x = datos["Celsius"]
y = datos["Fahrenheit"]
```

```python
y
```

```python
x
```

```python
#Para mirar el tipo de variable
#Podemos mirar un arreglo
x.values
```

```python
#Funcion de Numpy
x_procesada = x.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
#Creacion de nuestro modelo
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
#Entrenamiento
modelo.fit(x_procesada, y_procesada)
```

```python
#Crear una prediccion
celsius = 123
prediccion = modelo.predict([[celsius]])
print(f"{celsius} grados celsius son {prediccion} grados fahrenheit")
```

```python
#Poder hacer una evaluacion
modelo.score(x_procesada, y_procesada)
```

```python
#Para poder mirar que tipo de dato tiene la columna
'''
Podemos mirar una muestra de los datos que tenemos en nuesto csv
value_counts() = Nos dice todos los distintos datos que tiene y cuantos hay de cada uno

Nom_Variable["Columna que queremos mirar"].values_counts()
'''
datos["ocean_proximity"].value_counts()
```

```python
#Mirar la informacion de los datos
datos.info()
```

```python
#Funcion para analizar los datos
'''
Nos muestra la descripcion de las columnas
'''
datos.describe()
```

```python
#Creacion de graficas y destogramas
'''
Estamos creando graficas para mirar los datos de nuestras columnas del CSV
Nom_Variable.(hist) para crear las tablas ((figsize) Para agustar el formato de la imagen (bins)Para crear mas Barras (edgecolor)Para cambiar de color las barras)
'''
datos.hist(figsize=(15,8), bins=30, edgecolor ="black")
```

```python
#Importamos la libreria Seaborn(biblioteca de Python que permite crear grficos estadsticos atractivos y fciles de interpreta)
import seaborn as sb
#Colocamos las tablas que queramos mirar, asignando una variable para poder mostrarlo. Seguido de eso, colocamos el color
#Lo que creamos es para poder mirar que casas tienen mayor precio
sb.scatterplot(x="latitude", y="longitude", data=datos, hue="median_house_value", palette="coolwarm")
```

```python
#Para mirar, donde viven las personas que Tienen mas dinero
'''
Para eso en el apartado de:
    data=datos[(datos) (median_income / Nombre de la columna) > (Numero que queramos)]'''
sb.scatterplot(x="latitude", y="longitude", data=datos[(datos.median_income > 14)], hue="median_house_value", palette="coolwarm")
```

```python
#Mejorar el Set de datos
#Algunas columnas tienen datos basios, los vamos a quitar con la funcion (dropna) que significa: Quita los que tienen NA / Basios
#Nos mostrara datos nuevos donde quita todos los NA
datos_na = datos.dropna()

#Creamos una Nueva variable para poder aguardar, los nuevos datos. Donde quitamos los NA
```
